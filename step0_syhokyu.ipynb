ネコが 100 匹います。  
度数分布表を作成し、ヒストグラムを作成しなさい。


```python
# ネコの数
num_cats = 100

# ネコの身長データを生成 平均30cm,標準偏差4
cat_heights = np.random.normal(30, 4, num_cats)

# ヒストグラムと度数分布表の作成
hist, bin_edges = np.histogram(cat_heights, bins=[18, 22, 26, 30, 34, 38, 42])

# 度数分布表の表示
print("度数分布表:")
print("範囲\t\t度数")
for i in range(len(hist)):
    print(f"{bin_edges[i]:.1f}-{bin_edges[i+1]:.1f}\t{hist[i]}")

# ヒストグラムの作成と表示
plt.hist(cat_heights, bins=10, edgecolor="white")
plt.xlabel("(cm)")
plt.show()
```

7-1. バブルチャートを作ってみよう

```python
# データを作成
data = {
    'チーム': ['赤組', '青組', '黄組', '緑組', '白組'],
    'ボール投げ': [3.5, 7.5, 6, 4, 5],
    '50m走': [4, 4, 7, 7.5, 5],
    '高跳び': [8, 4, 6.5, 4.5, 5]
}

# データフレームを作成
df = pd.DataFrame(data)

# バブルチャートを作成
plt.figure(figsize=(4, 4))

for i in range(len(df)):
    plt.scatter(df.loc[i, 'ボール投げ'], df.loc[i, '50m走'], s=df.loc[i, '高跳び'] * 500, label=df.loc[i, 'チーム'])

# グラフにラベルを追加
plt.title('バブルチャート')
plt.xlabel('ボール投げ')
plt.ylabel('50m走')

# チーム名を各点に追加
for i in range(len(df)):
    plt.text(df.loc[i, 'ボール投げ'], df.loc[i, '50m走'], df.loc[i, 'チーム'], fontsize=10, ha='right')

# x軸およびy軸の目盛りを0から10に設定
plt.xlim(0, 10)
plt.ylim(0, 10)

# 凡例を表示
# plt.legend()

# グリッドを表示
plt.grid(True)

# グラフを表示
plt.show()
```
